# Exploratory Data Analysis (EDA)

```python
df['date_added']=df['date_added'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
df.head()
```

```python
df['loading_year'] = df['date_added'].dt.year
df['loading_year']
```

```python
df['loading_month'] = df['date_added'].dt.month
df['loading_month']
```

```python
df['type'].value_counts()
```

```python
pd.crosstab(df.type, df.loading_year)
```
